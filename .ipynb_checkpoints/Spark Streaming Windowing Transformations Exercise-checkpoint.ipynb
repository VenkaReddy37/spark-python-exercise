```python
from pyspark import SparkContext
from pyspark.streaming import StreamingContext
#Stoping SparkContext before starting another one
sc.stop()
# Create a local StreamingContext with two working thread and batch interval of 1 second
sc = SparkContext("local[2]", "Spark Streaming Windowing Transformations Exercise")
ssc = StreamingContext(sc, 10)
ssc.checkpoint("/Users/pravinkumar/Documents/Spark/Results/checkPointForStreaming")
```

```python
# Create a DStream that will connect to hostname:port, like localhost:9999
socket = ssc.socketTextStream("localhost", 9999)
```

```python
# Counting the words using reduceByKeyAndWindow
wordCountWindow = socket.map(lambda rec: rec.split("|")).\
flatMap(lambda rec: rec).map(lambda rec: (rec, 1)).\
reduceByKeyAndWindow(lambda acc, value: acc + value, 30, 10)

# Printing first 10 data samples of each batch of Dstream 
wordCountWindow.pprint()
```

```python
# Counting the words using window function 
wordCountWindow = socket.window(30, 10).map(lambda rec: rec.split("|")).\
flatMap(lambda rec: rec).map(lambda rec: (rec, 1)).\
reduceByKey(lambda acc, value: acc + value)

# Printing first 10 data samples of each batch of Dstream 
wordCountWindow.pprint()
```

```python
# Counting the words using countByValueAndWindow
wordCountWindow = socket.map(lambda rec: rec.split("|")).\
flatMap(lambda rec: rec).countByValueAndWindow(30, 10)

# Printing first 10 data samples of each batch of Dstream 
wordCountWindow.pprint()
```

```python
ssc.start()             # Start the computation
ssc.awaitTermination()  # Wait for the computation to terminate
```
