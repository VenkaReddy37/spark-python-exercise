```python
from pyspark import SparkContext
from pyspark.streaming import StreamingContext

sc.stop()

sc = SparkContext("local[*]", "SparkStreaming Join ")
ssc = StreamingContext(sc, 10)
ssc.checkpoint("/Users/pravinkumar/Documents/Spark/Results/checkPointForStreaming")
```

```python
socket01 = ssc.socketTextStream("localhost", 9000)

socket02 = ssc.socketTextStream("localhost", 8000)

DStream01 = socket01.map(lambda rec: rec.split("|")).map(lambda rec: (rec[0], rec[1]))
DStream01.pprint()
DStream02 = socket01.map(lambda rec: rec.split("|")).map(lambda rec: (rec[0], rec[2]))
DStream02.pprint()
StreamJoin = DStream01.join(DStream02)

StreamJoin.pprint()
StreamJoin.saveAsTextFiles("/Users/pravinkumar/Documents/Spark/Results/DeckStreamingJoinOp/Deck")
```

```python
# Test data
input1 = [[1, True], [2, False], [3, False], [4, True], [5, False]]
input2 = [[1, False], [2, False], [3, True], [4, True], [5, True]]

input1RDD = [sc.parallelize(input1)]
input2RDD = [sc.parallelize(input2)]

# Creates a DStream from the test data
input1DStream = ssc.queueStream(input1RDD).map(lambda rec: (rec[0], rec[1]))
input1DStream.pprint()
input2DStream = ssc.queueStream(input2RDD).map(lambda rec: (rec[0], rec[1]))
input2DStream.pprint()
# Join the two streams together by merging them into a single dstream
joinedDStream = input1DStream.join(input2DStream)
# Print the result 
joinedDStream.pprint()
```

```python
# Start the context, time out after one batch, and then stop it
ssc.start()
ssc.awaitTerminationOrTimeout(5000)
ssc.stop()
```
